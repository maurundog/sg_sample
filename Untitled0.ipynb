```
import tkinter #Tkクラスを含むライブラリを利用可能に
win = tkinter.Tk() #ウィンドウ作成
win.title("ようこそ") #タイトルを設定
win.minsize(640, 480) #ウィンドウサイズを設定
win.option_add("*font", ["MS Pゴシック", 22]) #フォントを設定
can = tkinter.Canvas(width=640, height=480, bg="white") #キャンバス作成
can.place(x=0, y=0) #キャンバスを設置
img = tkinter.PhotoImage(file="r18.png") #画像ファイルの読込
can.create_image(640 - 480 / 2, 480 - 260 / 2, image=img) #画像の設置
lab = tkinter.Label(text="教えてあげるね", bg="white") #ラベルの作成
lab.place(x=200, y=300) #ラベルを設置
win.mainloop() #メインループ
```

```
y = int(input("薬草の数は？"))
d = int(input("毒消の数は？"))
w = float(input("割引率は？"))
#関数の定義
def calc(yn, dn, wa):
    return (yn * 8 + dn * 10) * wa

print("薬草が{0}個、毒消が{1}個で{2}Gです".format(y, d, calc(y, d, w)))
```
